```python
from mplsoccer import Pitch, VerticalPitch
import matplotlib.pyplot as plt
import numpy as np


fig, axs = plt.subplots(2,2)

pitch = Pitch(axis=True, label=True)
pitch.draw(axs[0 , 0])
axs[0,0].set_title("Length")

pitch = Pitch(pitch_color="green")
pitch.draw(axs[0, 1])
axs[0,1].set_title("Pitch")
pitch = Pitch(axis=True, label=True)
pitch.draw(axs[1,0])

pitch = Pitch(positional=True, shade_middle=True)
pitch.draw(axs[1,1])
axs[1,1].set_xlabel("Length")

fig.tight_layout()

plt.show()
```
