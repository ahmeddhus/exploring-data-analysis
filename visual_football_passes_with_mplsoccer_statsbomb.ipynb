```python
from statsbombpy import sb
from mplsoccer import Pitch
```

```python
events = sb.events(match_id=22912)
events
```

```python
events.columns
```

```python
pass_events = events[['team', 'type', 'location', 'pass_end_location', 'player']]
pass_events = pass_events[pass_events['type'] == 'Pass']
pass_events.head(30)
```

```python
lp_pass_events = pass_events[pass_events['team'] == 'Liverpool']
lp_pass_events
```

```python
start_locations = lp_pass_events['location'].tolist()
start_locations
```

```python
end_locations = lp_pass_events['pass_end_location'].tolist()
end_locations
```

```python
start_x = [loc[0] for loc in start_locations]
start_y = [loc[1] for loc in start_locations]
end_x = [loc[0] for loc in end_locations]
end_y = [loc[1] for loc in end_locations]
start_x, start_y, end_x, end_y
```

```python
pitch = Pitch(pitch_color="green")
fig, ax = pitch.draw()

ax.scatter(start_x, start_y)
ax.scatter(end_x, end_y)

for i in range(len(start_x)):
    ax.plot([start_x[i], end_x[i]], [start_y[i], end_y[i]], color="yellow")
```
